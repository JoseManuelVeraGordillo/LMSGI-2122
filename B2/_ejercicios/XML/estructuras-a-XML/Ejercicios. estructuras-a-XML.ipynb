#### Ejercicio 1.

Convierte la siguiente estructura ...

```python
{
    "para": "Pedro",
    "de": "Laura",
    "titulo": "Recordatorio",
    "contenido": "A las 7:00 pm en la puerta del teatro"
}
```

en el siguiente fichero XML;

```xml
<nota>
   <para>Pedro</para>
   <de>Laura</de>
   <titulo>Recordatorio</titulo>
   <contenido>A las 7:00 pm en la puerta del teatro</contenido>
</nota>
```

````python
## Explicación para la resolución del Ejercicio-1
import xml.etree.ElementTree as et

# Para crear un XML en Python con el módulo XML.etree.ElementTree.. hay que realizar siempre una serie de pasos..

# HAY QUE CREAR LA ESTRUCTURA, PARTIENDO DE LA RAIZ, E IR AGREGANDO ELEMENTOS (NODOS) A LA RAIZ. 
# Crear un nodo (etiqueta XML)
#  a ese nodo, asignar nombre-etiqueta, valor (contenido) y atributos si los tuviera.
#  asignar ese nodo a un padre... todos los nodos tienen un padre excepto el Raiz.
#
#  ----> Para finalizar lo que queda es guardar ese arbol de nodos creados --- 
# ====================================================================================================

# VEAMOS UN EJEMPLO ==== 
root = et.Element("nota")  #Creo una etiqueta/nodo con nombre 'nota', y que será el nodo raiz == y será raiz porque no tiene un nodo PADRE.

# Ahora se crea los nodos de nota
tagPara = et.Element("para")
tagPara.text = "Pedro"  # Asigno el contenido a la etiqueta/nodo

root.append(tagPara)  #Asigno el nodo PADRE de la etiqueta '<para>' agregando a su padre, tagPara como hijo.

# repito este proceso con cada nodo
tagDe = et.Element("de")
tagDe.text = "Laura"  # Asigno el contenido a la etiqueta/nodo
root.append(tagDe)  

tagTitulo = et.Element("titulo")
tagTitulo.text = "LauRecordatoriora"  # Asigno el contenido a la etiqueta/nodo
root.append(tagTitulo)

tagContenido = et.Element("contenido")
tagContenido.text = "A las 7:00 pm en la puerta del teatro"  # Asigno el contenido a la etiqueta/nodo
root.append(tagContenido)


# Por último lo que queda es asignar el nodo RAIZ al XML (arbol) que quiero guardar
tree = et.ElementTree(root)

fichero = "nota.xml" 
tree.write(fichero)   #guardo el Arbol en XML indicando un parámetro que es el nombre del fichero.
````
